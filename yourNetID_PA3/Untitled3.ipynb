```python
import pandas as pd
import csv
wal = pd.read_csv('dataset_group.csv')
```

```python
wal.rename(columns={'2000-01-01':'Date', '1':'ID','yogurt':'Items'}, inplace=True)
wal.drop('Date', axis=1, inplace=True)
```

```python
wal.head()
```

```python
d=wal.groupby('ID').agg(lambda x: x.tolist())
  
```

```python
d
```

```python
#Code for Task-1
with open('dataset.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',', quotechar="'", quoting=csv.QUOTE_MINIMAL)
    for index, row in d.iterrows():
        writer.writerow(row[0])
csvfile.close()
```
