```python
%matplotlib inline
import os
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib.pyplot import hist
import seaborn as sns

df_olympics = pd.read_csv('athlete_events.csv', encoding='latin1')
df_olympics.head() 

Count_Row=df_olympics.shape[0]
Count_Col=df_olympics.shape[1]

print (">>Task 1-a: Details of df_olympics data frame are: \n",)
print(pd.DataFrame.info(df_olympics))

print ("\n\n>>Task 1-b: Number of rows:%s and number of columns:%s" % (Count_Row, Count_Col)) 

print ("\n\n>>Task 1-c: Descriptive details of age column are\n",df_olympics['Age'].describe())
result11 = df_olympics['Year'].max() - df_olympics['Year'].min()
num_uniq_games = None
print ("\n\n >>Task 1-d: In our dataset we have historical data for %s years, and for %s games. " % (result11, df_olympics['Games'].nunique()))
print("\n")
num_female_2004 = df_olympics[(df_olympics.Sex == 'M') & (df_olympics.Year == 2004)].shape[0]
num_male_2004   = df_olympics[(df_olympics.Sex == 'F') & (df_olympics.Year == 2004)].shape[0]
print (">>Task 2-a: At the Olympics of 2004, there were participating %s female and %s male athletes" 
       % (num_female_2004, num_male_2004))
print("\n")
num_medals_1986 = df_olympics[(df_olympics.Medal.notnull()) & (df_olympics.Year == 1896)].shape[0]
num_medals_2016 = df_olympics[(df_olympics.Medal.notnull()) & (df_olympics.Year == 2016)].shape[0]
print ("\n\n>>Task 2-b: The total number of metals awarded in 1896 was %s, while in 2016 was %s" 
       % (num_medals_1986, num_medals_2016))
print("\n")
grp = df_olympics[df_olympics['Season'] == 'Summer'].groupby('Year')['Name'].count()
print (grp)
sns.set_style('whitegrid')
sns.set(font_scale = 1.3)
grp.plot(kind='line') 
plt.show()
grp.plot(kind='bar') 
plt.show()
tallest1 = df_olympics.nlargest(10,'Height',keep='last')
print(tallest1)
```
